```python
import csv
x_val = []
y_val = []
z=0
repeat_value=5
k_range=range(1,26)
with open('iris.data', 'r') as csvfile:
	lines = csv.reader(csvfile)
	for row in lines:
            if z <=149:
                x_val.append(row[:4])
                y_val.append(row[4])
                z+=1
```

```python
import warnings
warnings.filterwarnings('ignore')
mydict={}
for step in range(1,repeat_value+1):
  from sklearn.model_selection import train_test_split
  x_train,x_test,y_train,y_test = train_test_split(x_val,y_val,test_size=0.33)
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn import metrics
  score = {}
  scoreList = []
  for k in k_range:
      neigh = KNeighborsClassifier(n_neighbors=k)
      neigh.fit(x_train,y_train)
      yPred = neigh.predict(x_test)
      score[k] = metrics.accuracy_score(y_test,yPred)
      scoreList.append(metrics.accuracy_score(y_test,yPred))
  
  mydict["step%s"%step]=scoreList

  num = 1
  for x in scoreList:
      print(num,"K: ",x)
      num+=1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
k_range=range(1,26)
for k,v in mydict.items():
  ax=plt.axes()
  scoreList=v
  print(scoreList)
  ax.plot(k_range,scoreList)
  ax.set(title = k ,xlabel="K Value",ylabel="Acc")
  plt.show()
```

```python
for j in  k_range:
  print("Average of K%s = %.2f" % (j,sum(v[j-1] for k , v in mydict.items()) / repeat_value))
```
